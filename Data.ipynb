```python
import matplotlib.pyplot as plt
import numpy as np

# Base HP at level 1 and level 60
base_hp_lvl_1 = [120, 120, 100, 100, 80, 80, 80, 120, 120]
base_hp_lvl_60 = [6000, 5500, 4000, 3500, 3000, 2800, 2800, 5000, 4500]

# Base HP + equipment at level 60
equip_hp_lvl_60 = [9000, 8500, 6000, 5000, 4500, 4300, 4300, 8000, 7500]

classes = ['Warrior', 'Paladin', 'Hunter', 'Rogue', 'Priest', 'Warlock', 'Mage', 'Shaman', 'Druid']
levels = np.arange(1, 61)

# Calculate HP per level for each class (without equipment)
base_hp_per_level = []
for i in range(len(classes)):
    hp_increase_per_level = (base_hp_lvl_60[i] - base_hp_lvl_1[i]) / 59
    base_hp_per_level.append(base_hp_lvl_1[i] + hp_increase_per_level * levels)

# Plot base HP per level for each class
plt.figure(figsize=(8, 4))
for i in range(len(classes)):
    plt.plot(levels, base_hp_per_level[i], label=classes[i])
plt.xlabel('Level')
plt.ylabel('Base HP')
plt.title('Base HP evolution per level for each class in WoW (2005)')
plt.legend(loc='upper left')
plt.show()

# Calculate HP per level for each class (with equipment)
equip_hp_per_level = []
for i in range(len(classes)):
    hp_increase_per_level = (equip_hp_lvl_60[i] - base_hp_lvl_60[i]) / 59
    equip_hp_per_level.append(base_hp_per_level[i] + hp_increase_per_level * levels)

# Plot base HP + equipment per level for each class
plt.figure(figsize=(8, 4))
for i in range(len(classes)):
    plt.plot(levels, equip_hp_per_level[i], label=classes[i])
plt.xlabel('Level')
plt.ylabel('Base HP + Equipment')
plt.title('Base HP + Equipment evolution per level for each class in WoW (2005)')
plt.legend(loc='upper left')
plt.show()

for i in range(len(classes)):
    m_base = (base_hp_lvl_60[i] - base_hp_lvl_1[i]) / 59
    m_equip = (equip_hp_lvl_60[i] - base_hp_lvl_60[i]) / 59
    m_total = m_base + m_equip
    b = base_hp_lvl_1[i]
    print(f'{classes[i]}: HP + Equip(x) = {m_total:.2f}x + {b}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Base HP at level 1 and level 60 (same as the previous example)
base_hp_lvl_1 = [120, 120, 100, 100, 80, 80, 80, 120, 120]
base_hp_lvl_60 = [6000, 5500, 4000, 3500, 3000, 2800, 2800, 5000, 4500]

classes = ['Warrior', 'Paladin', 'Hunter', 'Rogue', 'Priest', 'Warlock', 'Mage', 'Shaman', 'Druid']
levels = np.arange(1, 61)

# Calculate HP per level for each class (without equipment)
base_hp_per_level = []
for i in range(len(classes)):
    hp_increase_per_level = (base_hp_lvl_60[i] - base_hp_lvl_1[i]) / 59
    base_hp_per_level.append(base_hp_lvl_1[i] + hp_increase_per_level * levels)

# Calculate damage taken as a percentage of total HP for each class at each level
total_avg_damage = 600 * 4  # Increased damage due to reproduction rate
damage_percentage = []
for i in range(len(classes)):
    if base_hp_per_level[i][-1] > 0:
        damage_percentage.append((total_avg_damage / base_hp_per_level[i]) * 100)
    else:
        damage_percentage.append([0])

# Plot damage percentage per level for each class
plt.figure(figsize=(8, 4))
for i in range(len(classes)):
    plt.plot(levels, damage_percentage[i], label=classes[i])
plt.xlabel('Level')
plt.ylabel('Damage as a percentage of total HP')
plt.title('Corrupted Blood damage per level for each class in WoW (2005) without Equipment')
plt.axhline(y=100, color='r', linestyle='--', label='Survival threshold')
plt.legend()
plt.show()

# Calculate the survival threshold level for each class
survival_threshold = []
for i in range(len(classes)):
    survival_level = np.where(np.array(damage_percentage[i]) <= 100)[0][0]
    survival_threshold.append(survival_level)

print("Survival threshold levels:")
for i in range(len(classes)):
    print(f"{classes[i]}: Level {survival_threshold[i]}")
```
